```python
"""this follows work that I have done relating to the question of producing piecewise approximations of 
the function arctan. The goal is to produce an equation with the minimum (maximum) difference between 
the piecewise functiond and arctan"""
d = [0, 1, 3^.5, 7^.5, 15^.5, 31^.5, 63^.5, 127^.5]
m = [1, 1/2, 1/4, 1/8, 1/16, 1/32, 1/64, 1/128]
b = [0]
c = []

#all the given parameters are necessary to find the next c, b values
def next_vals(d_1, m_1, b_0, m_0, d_0):
    left_side = arctan(d_1) - m_1*d_1 - 2*b_0
    c_coef = 2*m_0 - m_1
    c_0 = find_root(c_coef*x - arctan(x) - left_side, d_0, d_1)
    b_1 = m_0*c_0 + b_0 - m_1*c_0
    return c_0, b_1
for i in range(len(d)-1):
    m_0 = m[i]; d_0 = d[i]; m_1 = m[i+1]; d_1 = d[i+1];
    b_0 = b[i]
    c_, b_ = next_vals(d_1, m_1, b_0, m_0, d_0)
    b.append(b_); c.append(c_);
c = [0] + c + [gp(pi/2-b[-1])/m[-1]]
```
